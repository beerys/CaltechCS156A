```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
```

```python
data_1 = [[0,0,2],[0,1,1],[0,1,1]]
data_2 = [[1,1],[1,0],[0,1]]
```

```python
ax.scatter(data_1[0],data_1[1],zs=data_1[2],c='r', label='+1')
ax.scatter(data_2[0],data_2[1],zs=data_2[2],c='b',label = '-1')
plt.title('Counterexample for the \n3d Perceptron')
plt.legend()
plt.savefig('plot_p4.png', dpi=100)
plt.show()
```
