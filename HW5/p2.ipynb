```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['lines.color'] = 'k'
mpl.rcParams['axes.prop_cycle'] = mpl.cycler('color', ['k'])
```

```python
x = np.linspace(-9, 9, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
```

```python
def axes():
    plt.axhline(0, alpha=.1)
    plt.axvline(0, alpha=.1)
```

```python
a = 2.
b = 1.
axes()
plt.contour(x, y,(x**2/a**2 - y**2/b**2), [1], colors='k')
plt.contour(x, y,(x**2/a**2 - y**2/b**2), [0.1], colors='b') #this is the positive sign portion
plt.contour(x, y,(x**2/a**2 - y**2/b**2), [5], colors='r') #this is the negative sign portion
plt.title('Positive and negative regions')
plt.xlabel('x_1')
plt.ylabel('x_2')
plt.legend()
plt.savefig('p2.jpg', dpi = 100)
plt.show()
```
